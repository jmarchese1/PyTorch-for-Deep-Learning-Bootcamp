# **Why would we want to use machine learning or deep learning?**

For a complex problem can you think of all the rules?

For example, self driving cars. There might be thousands of rules.

ML can be used for anything as long as you can convert it into to numbers and program it to find patterns. However....

*Googles #1 rule of machine learning: If you can build a simple rule based system that doesnt require machine learning, do that. - A wise software engineer. (just because its powerful, doesn't mean we should always use it)*

**what deep learning is good for**

problems with long lists of rules- when the traditional approach fails, machine learning / deep learning may help

continually changing enviorments - deep learning can adapt to new scenarios

Discovering insights within large collections of data

**what deep learning is not good for**

*typically*

When you need explainability - the patterns are typically uninterpretable by a human

when the traditional approach is a better option, a traditional based system is something like a recipe with 100% certainty of the output

when errors are unacceptable - since the outputs are always predictable they are probabalistic bets

when you dont have much data - deep learning models typically require lots of data. However, there are techniques to use little data and get great results.

**structured data vs unstructured data**

deep learning is best on unstructured data while traditional machine learning models are best for structured data.

traditional ml algorithms:
random forest, XGBoost, logistic regression, support vector machines, nearest neighbors

deep learning algorithms:
convolutional neural network, transfer learning etc

the deep in deep learning comes from having lots of layers. Such as resnet 152 which has 152 different layers.

Overall architecture talks about the input, hidden layers and the output layer. Patterns, weights, embedding, feature vectors all refer to similar things. This is what the hidden layers discover.

A neural network uses a combination of linear and non linear lines to draw patterns in our data.

**Types of learning**:

**supervised learning**

lots of data (inputs) and lots of examples of what the output should like. We have data and labels.

**unsupervised learning**

We only have the data, we dont have the labels. With photos of cats and dogs the model will find the fundamental patterns and group them. But it wouldn't know the difference between the two.

**transfer learning**

Taking the patterns learned on a dataset from one model and transfering it to another model.

**What is deep learning actually used for?**

things that deep learning is used for:

- youtube recommended videos
- google translate and google translates speech recognition system
-computer vision object detection
-natural language processing (spam detection on email inboxes, classification: what emails are spam 1 and emails that are not spam 0.)
